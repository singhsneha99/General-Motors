```python
import requests
import pandas as pd

# URL for the provided GM link
gm_results_url = "https://investor.gm.com/news-releases/news-release-details/gm-releases-2023-third-quarter-results"
```

```python
df = pd.read_html(gm_results_url, skiprows=4)[0]

# Set the header using row 4
df.columns = df.iloc[0]

# Drop the rows that were used as header
df = df.drop(4)

# Reset index
df = df.reset_index(drop=True)

# Display the cleaned DataFrame
print(df)
```

```python
df.shape
```

```python

# Rename columns for clarity
df.columns = ["($M) except where noted", "Q3 2023", "Q3 2022", "Change", "% Change", "NaN"]

# Drop unnecessary rows with NaN values
#df = df.dropna(subset=["NaN"])
df = df.iloc[:, :-1]

# Display the cleaned DataFrame
print("Cleaned DataFrame:")
print(df)
```

```python
df.columns = ["($M) except where noted", "Q3 2023", "Q3 2022", "Change", "% Change"]
df
```

```python
# Remove dollar signs, commas, and non-breaking space characters, and convert to numeric
df[['Q3 2023', 'Q3 2022', 'Change']] = df[['Q3 2023', 'Q3 2022', 'Change']].replace('$[\(\)%\xa0]', '', regex=True).replace('[\(\)%\xa0]', '', regex=True).astype(float)

# Remove parentheses and percentage sign from percentage columns and convert to numeric
df['% Change'] = df['% Change'].replace('[\(\)%\xa0]', '', regex=True).astype(float) / 100.0

# Display the cleaned DataFrame
print(df)
```
